Descriptive statistics: This statistics deals with the description of the data.
* It talks about the measure of central tendency (mean, median, mode)
* it talks about the measure of dispersion/variablity(spread) (range, variance, standard deviation, iqr, coefficient of variability)
* it also talks about the measure of position (quartiles, percentiles, deciles).
* it also talks about the measure of shape(distribution).

What is the variance of life Expectancy?

```python
df['lifeExp'].var()
```

what is the standard deviation?

```python
df['lifeExp'].std()
```

standard deviation is the square root of variance.
* Spread comes from μ−σ  to μ+σ. 
* so 59-12 to 59+12 i.e 47 to 71. i.e data lies between these range.
* low standard deviation means data is clustered along the mean.
* high standard deviation means data lies widely spread.

```python
What is the coefficient of variability?
```

```python
coff_of_var=(df['lifeExp'].std()/df['lifeExp'].mean())*100
coff_of_var
```

Since coefficient of variability have certain rules:
* <10%: low variation.
* 10-30%: moderate variation.
* greater than 30%: high variation.
--------
* But our coff_of_var lies in between 10-30% so, our data lies in moderate region. moderate variation.

find the skewness in the column life expectancy.

```python
df['lifeExp'].skew()
```

Skewness have certain rules:
* zero: normal.
* positive: tail lies on the right side.(mean>Median)
* negative: tail lies on the left side.(Mean<Median)
--------------
* from above, we see negative skewness that means our data is negatively skewed i.e tails extend on left side.

find the kurtosis.

```python
df['lifeExp'].kurtosis()
```

Kurtosis have certain rules:
* Mesokurtic: Normal or zero or equal to 3. moderate peak and tails.
* Leptokurtic: >0 or >3 i.e heavy outliers heavy tails and high peaks.
* Platykurtic: <0 or <3 i.e less outliers flat and fewer peaks.
---------------
* since, our data is negative i.e <3 i.e platykurtic.(less outliers and fewer peaks.)
