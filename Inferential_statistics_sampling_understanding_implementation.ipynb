```python
df=px.data.gapminder()
df
```

#### Inferential statistics: Branch of statistics that deals with the decisions and conclusions from the population of sample data.
* Here, in inferential statistics population refers to entire group and statistic referes to sample.

#### Simple random sampling:This is a simple random sample of 100 country-year observations. Each observation had equal probability of selection

```python
sample=df.sample(n=100,random_state=1)
sample
```

### Systematic Sampling:Pick a random starting point, then select every 20th student.Select every k-th observation after a random start. This ensures uniform coverage of the dataset but may introduce bias if data has patterns.

```python
k=20
system_sample=df.iloc[::k]
system_sample
```

#### Stratified Sampling: Each continent is proportionally represented, reducing sampling bias.Example: Population of 1000 students → strata by gender; Male: 600, Female: 400; Sample size: 100 → pick proportionally (60 males, 40 females)

```python
# keep only one year:
df_2007 = df[df['year'] == 2007]
```

```python
# take five countries from each continent
stratified_sample = df_2007.groupby('continent').apply(lambda x: x.sample(min(len(x), 5), random_state=1))
stratified_sample
```

```python
stratified_sample[['continent', 'country', 'lifeExp']]
```

#### Cluster Sampling: Entire continents are selected, reducing cost but increasing sampling error.Population: 10,000 students in 50 schools; Randomly select 10 schools → survey all students in those 10 schools

```python
select_continents=['Asia','Africa']
cluster_sample=df[df['continent'].isin(select_continents)]
cluster_sample
```

#### Multistage Sampling: Stage 1: Randomly select districts (cluster); Stage 2: Randomly select schools within districts (SRS or stratified);

* Here Step 1: Select continents
* Step 2: Select countries within continents
* Step 3: Select years
* Sampling is performed in stages, mimicking real-world surveys.

```python
# choose continents:
continents = df['continent'].sample(2, random_state=1)
continents
```

```python
# choose countries:
sample_countries=(df[df['continent'].isin(continents)].groupby('continent').sample(3,random_state=1))
```

```python
sample_countries
```

```python
# stage 3 choose year:
final_sample=sample_countries[sample_countries['year']==1982]
```

```python
final_sample
```

#### Sampling Distribution : Regardless of population distribution, the sampling distribution of the mean approaches normality.This demonstrates Central Limit Theorem

```python
df_2007 = df[df['year'] == 2007]
sample_mean=[df_2007['lifeExp'].sample(30).mean() for i in range(1,1000)]
```

```python
df_2007['lifeExp'].mean()
```

```python
np.mean(sample_mean),np.std(sample_mean)
```

By repeatedly sampling 30 countries, I obtained a sampling distribution of the mean. Its mean matches the population mean, and its standard deviation (standard error) is much smaller, showing that sample means are more stable and reliable.
